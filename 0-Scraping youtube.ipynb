# Vrai début du programme

```python
import pandas as pd
from youtubesearchpython import VideosSearch
from tube_dl import Youtube
from __future__ import unicode_literals
import youtube_dl
import os
from moviepy.editor import *
```

```python
df=pd.read_csv("data.csv")
```

```python
df_sample=df.sample(n = 10, replace = False) 
```

```python
df_sample[["name","loudness"]]
```

```python
list_video= list(df_sample["name"])
list_name_mp3=[]
```

```python
for titles in list_video:
    try:
        videosSearch = VideosSearch(titles, limit = 2)
        result=videosSearch.result()
        url=result['result'][0]["link"]
        namemp3=result['result'][0]['title']+"-"+result['result'][0]['id']+".mp3"
        list_name_mp3.append(namemp3)
        namemp4=result['result'][0]['title']+"-"+result['result'][0]['id']+".mp4"
        #download video
        ydl_opts = {}
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        #convert mp4 to mp3
        video = VideoFileClip(os.path.join(namemp4))
        video.audio.write_audiofile(os.path.join(namemp3))
        os.system('ffmpeg -i {} -acodec pcm_s16le -ar 16000 {}/{}.wav'.format(namemp4, namemp4+'.wav', namemp4))
    except:
        pass
    
```
